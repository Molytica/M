```python
import sqlite3
import os
print(os.getcwd())

def connect_to_database():
    conn = sqlite3.connect('../../data/curated_chembl/SMILES_metadata.db')
    return conn

# Test the connection
conn = connect_to_database()

# Select the first 10 rows from the table mol_metadata
cursor = conn.cursor()
cursor.execute("SELECT * FROM mol_metadata LIMIT 10;")
rows = cursor.fetchall()

# Print the selected rows
for row in rows:
    print(row)
```
